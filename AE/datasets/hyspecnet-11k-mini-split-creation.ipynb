```python
import glob
import numpy as np
import os
import random
import shutil

random.seed(10587)
```

```python
in_directory = "./hyspecnet-11k/patches/"

out_directory = "./hyspecnet-11k/splits/mini/"
```

```python
train_percent = 0.7
val_percent = 0.2
test_percent = 0.1
```

```python
in_tiles_folders = glob.glob(f"{in_directory}**/")

random.shuffle(in_tiles_folders)
```

```python
num_tiles = len(in_tiles_folders)
```

```python
train_tiles_mini, val_tiles_mini, test_tiles_mini = np.split(in_tiles_folders, [int(num_tiles*train_percent), int(num_tiles*(train_percent+val_percent)+1)])
```

```python
train_patches_mini = []

for tile in train_tiles_mini:
    patch_npys = glob.glob(f"{tile}**/*DATA.npy")
    patch_npys.sort()
    center_patch_npy = patch_npys[len(patch_npys)//2]
    train_patches_mini.append(center_patch_npy)

train_patches_mini = [("/").join(patch.split("/")[-3:]) for patch in train_patches_mini]
```

```python
val_patches_mini = []

for tile in val_tiles_mini:
    patch_npys = glob.glob(f"{tile}**/*DATA.npy")
    patch_npys.sort()
    center_patch_npy = patch_npys[len(patch_npys)//2]
    val_patches_mini.append(center_patch_npy)

val_patches_mini = [("/").join(patch.split("/")[-3:]) for patch in val_patches_mini]
```

```python
test_patches_mini = []

for tile in test_tiles_mini:
    patch_npys = glob.glob(f"{tile}**/*DATA.npy")
    patch_npys.sort()
    center_patch_npy = patch_npys[len(patch_npys)//2]
    test_patches_mini.append(center_patch_npy)

test_patches_mini = [("/").join(patch.split("/")[-3:]) for patch in test_patches_mini]
```

```python
train_patches_mini.sort()
val_patches_mini.sort()
test_patches_mini.sort()
```

```python
np.savetxt(f"{out_directory}/train.csv", train_patches_mini, delimiter =", ", fmt ='% s')
np.savetxt(f"{out_directory}/val.csv", val_patches_mini, delimiter =", ", fmt ='% s')
np.savetxt(f"{out_directory}/test.csv", test_patches_mini, delimiter =", ", fmt ='% s')
```
